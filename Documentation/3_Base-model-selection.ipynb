# Base model selection

Within the base model selection, we decided to go with the Llama-2-series of meta as it currently consists of the best open source models available. Thereby, we selected the chat-series as we want our chatbot to be optimized for chat-use-cases at the end as well. This led us to the choice between the Llama-2-7b-chat-hf, Llama-2-13b-chat-hf and Llama-2-70b-chat-hf model.

## Evaluation of the Llama-2-model
After we downloaded the different Llama-2-models, we started with the evaluation of these. Therefore, we first loaded the different Llama-2-models into the text-generation-webui and then conducted a perplexity-evaluation and a loss-evaluation.

### Loading Llama-2-models

We used the following commands in the terminal to start the webui with the different models loaded in 8bit:

    conda activate textgen
    cd /home/[...]/text-generation-webui/
    python server.py --share --model models/Llama-2-7b-chat-hf --load-in-8bit
    python server.py --share --model models/Llama-2-13b-chat-hf --load-in-8bit
    python server.py --share --model models/Llama-2-70b-chat-hf --load-in-8bit


### Perplexity Evaluation
Afterwards, we started the Perplexity-Evaluation. Therefore, we conducted the following steps in the webui:

1. Go to the tab "Training"
2. Switch to the tab "Perplexity evaluation"
3. Select the dataset, you want to evaluate on; in our case: BPM_Evaluation_dataset.txt and BPMN_Evaluation_dataset.txt
4. The results of the evaluation, will be added to the table below (Corresponding CSV-File: "/home/[...]/text-generation-webui/logs/evaluation.csv")

Sidenote: One bug had to be fixed within the models before starting the evaluation. The following adaption has been made in the config.json file of the Llama-2-13b-chat-hf and Llama-2-70b-chat-hf model: pretraining_tp = 1.

### Loss Evaluation
The following steps were conducted in the webui for the loss evaluation:

1. Go to the tab "Training"
2. Stay in the tab "Train LoRA"
3. Change the Epochs-parameter to 1
4. Select the following files within the tab "Formatted Dataset"
    - Data-Format: Alpaca Format
    - Dataset: Structured_empty_dataset
    - Evaluation Dataset: Structured_evaluaton_dataset
5. The results of the evaluation will be shown in the terminal under "Evaluation_loss"

## Results of the Evaluation

After conducting the perplexity-evaluation and loss-evaluation on the different models, we received the following results:



As expected, the Llama-2-70b-chat-hf model demonstrated superior or at least similar performance compared to its smaller counterparts across all evaluated categories, owing to its substantially larger parameter count, while conversely, the Llama-2-7b-chat-hf model exhibited comparatively weaker performance across the same categories due to its limited parameter count. Notably, the performance gap between the Llama-2-7b-chat-hf model and the Llama-2-13b-chat-hf model was more pronounced than the difference observed between the latter and the Llama-2-70b-chat-hf model. This discrepancy is noteworthy, especially considering that the parameter disparity between the first and second models is considerably smaller than that between the second and third model. Additionally, it is imperative to acknowledge the inherent challenges associated with training and operating a model of such monumental proportions as the 70B-parameter variant. Consequently, to balance model capacity, practical manageability, and quality, we opted for the Llama-2-13b-chat-hf-model as the most suitable base model.
