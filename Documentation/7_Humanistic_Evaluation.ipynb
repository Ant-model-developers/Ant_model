# Humanistic Evaluation [EVAL4]
The humanistic evaluation is not conducted in the text-generation-webui, but instead uses a newly created gradio app. This gradio app can be found in the folder Humanistic_Evaluation under the file Humanistic_evaluation.py. The following document describes how the gradio app can be started, how the evaluation was conducted and shows the results of the humanistic evaluation for the order-to-cash process. The XML-Code of the order-to-cash process can be found in the folder "Humanistic_Evaluation/Example_processes/".

## Start gradio app
To start the gradio app, the following commands were executed in the terminal:

    conda activate evaluation
    cd /home/[...]/Humanistic_Evaluation/
    python Humanistic_evaluation.py
    
Then the public url was copied into the browser to access the webui.
    
## Evaluation
Within the webui, we copied the XML-Code of the Order-to-Cash process into the prompt and started asking the chatbot questions about BPM and BPMN in general and question about the process in specific. The first questions regarding the process tested if the chatbot in general understood the XML-Code of the process and can explain it in natural language. Therefore, we asked it to summarize the process and asked him to answer logical questions about the process. After we ensured that the chatbot understood the process correctly, we continued to the steps "Process Analysis" and "Process Redesign" of the BPM Lifecycle. Within the Process Analysis step, we asked the chatbot about possible strengths and weaknesses of the process as well as about the possible consequences of these weaknesses. This led us to the next step - the process redesign. Within this phase, we asked the chatbot to propose solutions and redesign recommendations for the identified problems. The recommendations should thereby be on a conceptual level as well as on a specific level, where the chatbot should propose in specific how the BPMN-process should be changed.

## Results

Firstly, it has to be noted that the current Ant-model doesn't understands the process in 100% of the time correctly. Thereby, it sometimes produces either inconvenient or even wrong results. Sometimes, however, it als produces very good results. Thereby, we proved that such a technology can be of high value for organizations. We leave it open for future research to develop a chatbot that always produces reliable results. But as we only wanted to showcase a proof of concept, the following evaluation shows the results that were satisfactory for us.

### BPM and BPMN Knowledge [DO1] 

**User**: What is a process? 

**Chatbot**: A process is a sequence of activities that produce specific outcomes. It is a series of actions or events that are linked and ordered in a way that makes sense for a particular purpose. 

A process starts with a trigger event, which sets the process in motion. The trigger event is followed by a sequence of activities that perform a specific function, and the process ends when the desired outcome is achieved. 

There are different types of processes, such as: 

1. Task processes: These are the most common type of process and are used to perform a specific task or set of tasks. 

2. Event processes: These are triggered by a specific event and are used to handle that event. 

3. Message processes: These are used to send and receive messages. 

4. Compensation processes: These are used to undo the effects of a previous process. 

Each step in a process has a specific purpose and is linked to the following step through a condition or trigger. The condition or trigger defines the outcome of the step and determines which step will come next. 

For example, a process might start with a trigger event of "Receive order," which is followed by a step of "Fill order." If the order can be filled, the process would continue to the step of "Deliver order." If the order cannot be filled, the process would go back to the step of "Fill order" to try again. 

Overall, processes are a way of automating repetitive tasks and providing a structured way of completing a specific task or set of tasks.

BPM can help organizations: 

1. Streamline processes: Automate repetitive tasks and reduce manual tasks to increase productivity. 

2. Improve process efficiency: Eliminate waste, reduce cycle times, and increase throughput. 

3. Enhance customer satisfaction: Improve process agility and responsiveness to customer needs.

6. Improve compliance: Ensure adherence to regulatory requirements and industry standards. 

BPM includes a set of tools and techniques, such as: 

1. Business Process Model and Notation (BPMN): A standardized graphical representation of business processes. 

2. Workflow management systems: Software that automates business processes and manages workflows. 

3. Process automation: Technology that automates repetitive tasks and processes. 

4. Process monitoring and analytics: Tools that track and measure process performance. 

5. Process optimization: Techniques for improving process efficiency and effectiveness. 

Overall, BPM is a discipline that helps organizations manage and improve their business processes to achieve strategic objectives and gain a competitive advantage.

### Process Redesign [DO3] 

**User**: How would you redesign the process to tackle the weaknesses? 

**Chatbot**: To address the weaknesses in the order processing process, I would suggest the following redesign: 


1. Automate the order processing: 

- Implement an order management system that can handle the entire process, from receiving the order to fulfilling it, and provide visibility into the status of orders. 

- Use APIs or other integration techniques to automate the communication between the order management system and other systems, such as the CRM or inventory management systems. 

2. Improve the customer experience: 

- Provide customers with real-time updates on the status of their orders through email, SMS, or a customer portal. 

- Offer multiple channels for customers to place orders, such as online, by phone, or through a mobile app. 

3. Streamline order fulfillment: 

- Implement an inventory management system that can integrate with the order management system and provide up-to-date information on inventory levels. 

- Use barcode scanning or RFID technology to automate the picking and packing process, and to provide real-time updates on the status of orders. 

- Consider introducing a "pick-to-light" or "pick-to-voice" system to further streamline the order fulfillment process. 

4. Provide clear escalation procedures: 

- Define clear escalation procedures for orders that require special attention, such as issues with inventory or shipping, and ensure that these procedures are communicated to all relevant parties. 

5. Implement metrics and KPIs: 

- Define and implement metrics and KPIs to measure the performance of the order processing process, such as cycle time, lead time, and defect rate. 

- Use these metrics to identify areas for improvement and to monitor progress over time. 

6. Improve communication and collaboration: 

- Implement a communication and collaboration platform that allows all relevant parties to share information and work together on orders. 

- Use this platform to facilitate communication and collaboration between the sales representative, order fulfillment team, and customer service team. 


By implementing these changes, the order processing process can be redesigned to address the identified weaknesses and improve overall efficiency and customer satisfaction.
