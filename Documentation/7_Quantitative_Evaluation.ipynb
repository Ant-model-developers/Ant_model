# Quantitative Evaluation [EVAL3]
The quantitative evaluation consists of the unsupervised perplexity evaluation and the supervised loss evaluation. Both evaluations take place within the text-generation-webui. Thereby, we compare our ProcessLLM to the Llama-2-chat-models, the intermediate model, the Alpaca-7b and 13b-models, the Vicu√±a-7b- and 13b-models, the Guanaco-13b-model and the MPT-7b- and 30b-chat-models. Therefore, the new models were as well downloaded into the folder "/home/[...]/text-generation-webui/models/".

The following notebook describes examplary how the perplexity evaluation and the loss evaluation were conducted for the ProcessLLM as the evaluations for the other models were conducted identically. The results of the evaluations can be found at the end of this notebook.


## Loading model into text-generation-webui
We used the following commands in the terminal to start the webui and to load the ProcessLLM in 8bit into it:

    conda activate textgen
    cd /home/[...]/text-generation-webui/
    python server.py --share --model models/ProcessLLM --load-in-8bit
    
## Perplexity Evaluation
Afterwards, we started the Perplexity-Evaluation. Therefore, we conducted the following steps in the webui:

1. Go to the tab "Training"
2. Switch to the tab "Perplexity evaluation"
3. Select the dataset, you want to evaluate on; in our case: BPM_Evaluation_dataset.txt and BPMN_Evaluation_dataset.txt
4. The results of the evaluation, will be added to the table below (Corresponding CSV-File: "/home/[...]/text-generation-webui/logs/evaluation.csv")

## Loss Evaluation
The following steps were conducted in the webui for the loss evaluation:

1. Go to the tab "Training"
2. Stay in the tab "Train LoRA"
3. Change the Epochs-parameter to 1
4. Select the following files within the tab "Formatted Dataset"
    - Data-Format: Alpaca Format
    - Dataset: Structured_empty_dataset
    - Evaluation Dataset: Structured_evaluation_dataset
5. The results of the evaluation will be shown in the terminal under "Evaluation_loss"
